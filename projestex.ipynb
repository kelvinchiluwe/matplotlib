```python
#4
df=pd.read_csv('historical weather data.csv')
df['dt_iso']=pd.to_datetime(df['dt_iso'])
# df['year'] = df['dt_iso'].dt.year
# df['month'] = df['dt_iso'].dt.month

df['dt_iso']
```

```python
import matplotlib.pyplot as plt

# Compute the monthly average temperature
monthly_averageTemp = df.groupby('city_name')['temp'].mean()

# Extract x and y values for plotting
x = monthly_averageTemp.index  # City names
y = monthly_averageTemp.values  # Average temperatures

# Create the line plot
plt.plot(x, y, color='blue', marker='*', linestyle='-')  # Line graph with markers
plt.xlabel('City Name')
plt.ylabel('Monthly Average Temperature')
plt.title('Monthly Average Temperature for Each City')
plt.xticks(rotation=45)  # Rotate city names for readability
plt.grid(True)

plt.show()
```
